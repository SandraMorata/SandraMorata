### 3. Matriz inversa
Fija la semilla de aleatoriedad a 0, y genera una matriz aleatoria de enteros del 0 al 10 de tamaño 5x5.
Después de esto, calcula la inversa de dicha matriz.

```python
# @title
np.random.seed(0)
matriz_random = np.random.randint(0, high=11, size=5*5, dtype=int).reshape((5,5))
matriz_random_inv = np.linalg.inv(matriz_random)

np.matmul(matriz_random, matriz_random_inv)
```

### 4. Multiplicación matricial
Ahora multiplica tu matriz aleatoria original por su inversa. El resultado... ¿Es una matriz especial?

PISTA: Números muy pequeños se pueden entender como que son 0.

```python
# @title
resultado = np.matmul(matriz_random, matriz_random_inv)
print(resultado) # Es la matriz Identidad
```

- Usando NumPy.

Luego, compara el tiempo de ejecución para 10000 cálculos

```python
# @title
array_x1 = np.arange(1,10001)
array_x2 = np.arange(1,10001) * 2 + 1
array_y1 = np.arange(1,10001) * 3 - 1
array_y2 = np.arange(1,10001) - 1

lista_x1 = array_x1.tolist()
lista_x2 = array_x2.tolist()
lista_y1 = array_y1.tolist()
lista_y2 = array_y2.tolist()
```
